```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data from your Excel file into a Pandas DataFrame
df = pd.read_csv('StudentsPerformance.csv')
```

<h3> Find out how many males and females participated in the test.</h3>

```python
# Count the number of males and females
male_count = df[df['gender'] == 'male'].shape[0]
female_count = df[df['gender'] == 'female'].shape[0]
# Print the results
print("Number of Males: ", male_count)
print("Number of Females: ", female_count)
```

<h3> What do you think about the students' parental level of education? </h3>

```python
# Count the number of students at each level of parental education
education_counts = df['parental level of education'].value_counts()

# Print the results
print(education_counts)
```

```python
# Create a bar chart of the education counts
education_counts.plot(kind='bar')

# Add chart title and axis labels
plt.title('Parental Level of Education')
plt.xlabel('Education Level')
plt.ylabel('Number of Students')

# Show the chart
plt.show()
```

<h3> Who scores the most on average for math, reading and writing based on Gender </h3>

```python
# Calculate the average score for each subject based on gender
gender_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

# Print the results
print(gender_scores)
```

```python
max_math_score_gender = gender_scores['math score'].idxmax()
print("Gender with the highest average math score: ", max_math_score_gender)

max_read_score_gender = gender_scores['reading score'].idxmax()
print("Gender with the highest average reading score: ", max_read_score_gender)

max_write_score_gender = gender_scores['writing score'].idxmax()
print("Gender with the highest average writing score: ", max_write_score_gender)
```

<h3> Who scores the most on average for math, reading and writing based on Test preparation course </h3>

```python
# Calculate the average score for each subject based on test preparation course
prep_scores = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()

# Print the results
print(prep_scores)
```

```python
max_math_score_prep = prep_scores['math score'].idxmax()
print("Test preparation course with the highest average math score: ", max_math_score_prep)

max_read_score_prep = prep_scores['reading score'].idxmax()
print("Test preparation course with the highest average reading score: ", max_read_score_prep)


max_write_score_prep = prep_scores['writing score'].idxmax()
print("Test preparation course with the highest average writing score: ", max_write_score_prep)
```

<h3> What do you think about the scoring variation for math, reading and writing
based on Gender. </h3>

```python
# Calculate the variance and standard deviation for each subject based on gender
gender_variances = df.groupby('gender')[['math score', 'reading score', 'writing score']].var()
gender_std_devs = df.groupby('gender')[['math score', 'reading score', 'writing score']].std()

# Print the results
print("Variance for each subject based on gender:\n", gender_variances)
print("\nStandard deviation for each subject based on gender:\n", gender_std_devs)
```

```python
# Calculate the variance score for each subject and gender
male_var = df[df['gender'] == 'male'][['math score', 'reading score', 'writing score']].var()
female_var = df[df['gender'] == 'female'][['math score', 'reading score', 'writing score']].var()

# Create a bar chart showing the variance score for each subject and gender
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = ['Math', 'Reading', 'Writing']
male_bars = ax.bar([i - bar_width/2 for i in range(len(index))], male_var, bar_width, label='Male')
female_bars = ax.bar([i + bar_width/2 for i in range(len(index))], female_var, bar_width, label='Female')

# Add labels, title, and legend
ax.set_xlabel('Subject')
ax.set_ylabel('Variance Score')
ax.set_title('Variance Math, Reading, and Writing Scores by Gender')
ax.set_xticks(range(len(index)))
ax.set_xticklabels(index)
ax.legend()

# Show the plot
plt.show()
```

```python
# Calculate the standard deviation score for each subject and gender
male_std = df[df['gender'] == 'male'][['math score', 'reading score', 'writing score']].std()
female_std = df[df['gender'] == 'female'][['math score', 'reading score', 'writing score']].std()

# Create a bar chart showing the variance score for each subject and gender
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = ['Math', 'Reading', 'Writing']
male_bars = ax.bar([i - bar_width/2 for i in range(len(index))], male_std, bar_width, label='Male')
female_bars = ax.bar([i + bar_width/2 for i in range(len(index))], female_std, bar_width, label='Female')

# Add labels, title, and legend
ax.set_xlabel('Subject')
ax.set_ylabel('Standard Deviation Score')
ax.set_title('Standard Deviation Math, Reading, and Writing Scores by Gender')
ax.set_xticks(range(len(index)))
ax.set_xticklabels(index)
ax.legend()

# Show the plot
plt.show()
```

<h3> What do you think about the scoring variation for math, reading and writing
based on Test preparation course </h3>

```python
# Calculate the variance and standard deviation for each subject based on test preparation course
prep_variances = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].var()
prep_std_devs = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].std()

# Print the results
print("Variance for each subject based on gender: \n", prep_variances)
print("\nStandard deviation for each subject based on gender:\n", prep_std_devs)
```

```python
# Calculate the variance score for each subject and test preparation course
com_var = df[df['test preparation course'] == 'completed'][['math score', 'reading score', 'writing score']].var()
non_var = df[df['test preparation course'] == 'none'][['math score', 'reading score', 'writing score']].var()

# Create a bar chart showing the variance score for each subject and test preparation course
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = ['Math', 'Reading', 'Writing']
com_bars = ax.bar([i - bar_width/2 for i in range(len(index))], com_var, bar_width, label='Completed')
non_bars = ax.bar([i + bar_width/2 for i in range(len(index))], non_var, bar_width, label='None')

# Add labels, title, and legend
ax.set_xlabel('Subject')
ax.set_ylabel('Variance Score')
ax.set_title('Variance Math, Reading, and Writing Scores by Test Preparation Course')
ax.set_xticks(range(len(index)))
ax.set_xticklabels(index)
ax.legend()

# Show the plot
plt.show()
```

```python
# Calculate the Standard Deviation score for each subject and test preparation course
com_std = df[df['test preparation course'] == 'completed'][['math score', 'reading score', 'writing score']].std()
non_std = df[df['test preparation course'] == 'none'][['math score', 'reading score', 'writing score']].std()

# Create a bar chart showing the Standard Deviation score for each subject and test preparation course
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = ['Math', 'Reading', 'Writing']
com_bars = ax.bar([i - bar_width/2 for i in range(len(index))], com_std, bar_width, label='Completed')
non_bars = ax.bar([i + bar_width/2 for i in range(len(index))], non_std, bar_width, label='None')

# Add labels, title, and legend
ax.set_xlabel('Subject')
ax.set_ylabel('Standard Deviation Score')
ax.set_title('Standard Deviation Math, Reading, and Writing Scores by Test Preparation Course')
ax.set_xticks(range(len(index)))
ax.set_xticklabels(index)
ax.legend()

# Show the plot
plt.show()
```

<h3> The management needs your help to give bonus points to the top 25% of students based on their math score, so how will you help the management to achieve this. </h3>

```python
# Sort the data by math score in descending order
sorted_data = df.sort_values(by='math score', ascending=False)

# Calculate the total number of students in the dataset
total_students = len(df)

# Calculate the number of students who will receive bonus points
num_bonus_students = int(round((25/100)*total_students))
#print(num_bonus_students)

# Select the top num_bonus_students students from the sorted data and give them bonus points
bonus_points = 20
top_students = sorted_data.head(num_bonus_students).copy()

# Print top_students before incrementing math score
print('Top students before incrementing math score:\n', top_students['math score'])

# Increment math score for top_students
top_students.loc[:,'math score'] += bonus_points

# Print top_students after incrementing math score
print('Top students after incrementing math score:\n', top_students['math score'])
```
